#1.1 Функции

```python
def read_xls(paper_name):
  df = pd.read_excel('/content/drive/MyDrive/Diploma/df.xlsx',index_col=0, sheet_name=paper_name)
  df = df.astype(float, errors='ignore')
  df.reset_index(inplace=True)
  return df
```

```python
def rename_columns(df, start_year=None, end_year=None):
    if len(df.columns) == 2:
        new_column_names = ['country', 'total'] #специально для таблицы df_mig
    else:
        new_column_names = ['country'] + [str(x) for x in range(start_year, end_year + 1)][1:] + ['total']

    df.columns = new_column_names
    return df
```

```python
def index_method(x, denominator, k):
    x_index = (x / denominator) * k
    result = "{:.4f}".format(x_index)
    return result
```

#1.4 Формула эмпирической верификации

$$_{k}^{1.5}i_{ЭВ} = _{k}^{0.35}i_{ТО} + _{k}^{0.35}i_{ПИИ} + _{k}^{0.32}i_{ИИ}+ _{k}^{0.15}i_{ПС} + _{k}^{0.11}i_{ЕС} + _{k}^{0.11}i_{КС} + _{k}^{0.11}i_{М}$$


$i_{ТО}$ – индекс торгового оборота;

$i=1$ – переменная, которая принимает в себя значение от 1 до n (количество стран);

$x_(i )$– переменная, включающая в себя перечень стран;

$cum x_i$    – кумулированный торговый оборот страны $i$ с Кыргызстаном;

$\sum_{i=1}^n$  – сумма торгового оборота всех стран с Кыргызстаном

$k$ - весовой коэффициент отражает долю каждого показателя в общем наборе данных. в нашем сравнительном индексе позволяет учитывать различную важность каждого из показателей при расчете общего индекса


пример на формуле торгового оборота

$$i_{ТО} = \sum_{i=1}^n{\left(\frac{\text{cum }x_i \cdot k^{0.35}}{\sum_{i=1}^n }\right)}$$

Примечание: собственная разработка автора

#2.1 Индекс торговли

```python
df_trade['2002-2011'] = df_trade.iloc[:, 1:11].fillna(0).sum(axis=1) # разделяем df на части
df_trade['2012-2021'] = df_trade.iloc[:, 11:21].fillna(0).sum(axis=1)
df_trade['total'] = df_trade.iloc[:, 21:].fillna(0).sum(axis=1)

df_trade_indexed = df_trade[['country','2002-2011','2012-2021','total']]

denominator_2011 = df_trade_indexed.iloc[:6]['2002-2011'].sum()
denominator_2021 = df_trade_indexed.iloc[:6]['2012-2021'].sum()
denominator_total = df_trade_indexed.iloc[:6]['total'].sum()

df_trade_indexed['2002-2011_index'] = df_trade_indexed['2002-2011'].apply(lambda x: index_method(x, denominator_2011, 0.35))
df_trade_indexed['2012-2021_index'] = df_trade_indexed['2012-2021'].apply(lambda x: index_method(x, denominator_2021, 0.35))
df_trade_indexed['total_index'] = df_trade_indexed['total'].apply(lambda x: index_method(x, denominator_total, 0.35))
```

```python
df_trade_i = pd.merge(df_trade_indexed[['country']], df_trade_indexed.iloc[:, 4:], left_index=True, right_index=True)
```

#4. Финальная таблица индексов

```python
df_index.iloc[:,1:] = df_index.iloc[:,1:].astype(float)
```

```python
columns_2002_2011 = [("trade", "2002-2011"),
("FDI", "2004-2013"),
("DI", "2004-2013"),
("POL", "2015"),
("ECO", "2015"),
("CUL", "2015"),
("Migration", "total")]
```

```python
columns_2012_2021 = [("trade", "2012-2021"),
("FDI", "2014-2021"),
("DI", "2014-2021"),
("POL", "2017"),
("ECO", "2017"),
("CUL", "2017"),
("Migration", "total")]
```

```python
df_index['decade_1'] = df_index[columns_2002_2011].sum(axis=1) #суммируем колонны за первое десятилетие
df_index['decade_2'] = df_index[columns_2012_2021].sum(axis=1) #суммируем колонны за второе десятилетие
```

```python
#делаем метрики
```

```python
df_index['index_growth'] = df_index['decade_2'] - df_index['decade_1'] #измеряем рост
df_index['percentage_growth'] = df_index['decade_2'] / df_index['decade_1']*100 #измеряем рост в %
```

```python
def percentage_edit(row):
  if row < 100:
    return row - 100
  else:
    return row
```

```python
df_index['percentage_growth'] = df_index['percentage_growth'].apply(percentage_edit)
```
